# Linux Commands for the Review

## File and Directory Management
1. **`ls`**  
   - Lists files and directories.  
   - **Options:**  
     - `ls -l`: Detailed listing.  
     - `ls -ltr`: Detailed listing with reverse chronological order.  

2. **`touch [filename]`**  
   - Creates an empty file.  

3. **`cat [filename]`**  
   - Displays the contents of a file.  

4. **`pwd`**  
   - Prints the current working directory.  

5. **`cd [directory]`**  
   - Changes the current directory.  
   - `cd ..`: Moves up one directory level.  

6. **`mkdir [directory]`**  
   - Creates a new directory.  

7. **`rm [file or directory]`**  
   - Removes files or directories.  
   - **Options:**  
     - `rm -rf [directory]`: Removes a directory and its contents forcefully.  

## File Permissions
1. **`chmod [permissions] [filename]`**  
   - Changes file permissions.  
   - **Permission Breakdown:**  
     - `4`: Read.  
     - `2`: Write.  
     - `1`: Execute.  
   - **Examples:**  
     - `chmod 777 file`: Full permissions.  
     - `chmod 444 file`: Read-only for all.  

## Manual and History
1. **`man [command]`**  
   - Displays the manual for a command.  

2. **`history`**  
   - Lists previously executed commands.  

## Text Editors
1. **`vi [filename]`** / **`vim [filename]`**  
   - Opens the file in the editor.  
   - **Usage:**  
     - `i`: Insert mode.  
     - `Esc :wq`: Save and exit.  
     - `Esc :q!`: Quit without saving.  

## System Monitoring and Resource Usage
1. **`top`**  
   - Monitors system performance, showing CPU and memory usage.  

2. **`free`**  
   - Displays memory usage details.  

3. **`nproc`**  
   - Displays the number of CPUs available.
